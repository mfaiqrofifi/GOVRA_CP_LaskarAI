# **Menyatukan FIle**

```python
df1 = pd.read_csv("instagram_comments.csv")
df2 = pd.read_csv("dataset_tiktok-scraper_2025-05-18.csv")
df3 = pd.read_csv("dataset_tiktok-scraper_2025-05-15_14-45-42-854.csv")
df4 = pd.read_csv("Indonesian_Sentiment_Twitter_Dataset_Labeled.csv", on_bad_lines='skip')
df5 = pd.read_csv("review_layanan_jabar.csv")
```

```python
print(df4.columns)
```

```python
# Instagram: ambil kolom 'comment' atau 'posts'
df1_text = df1[['comment']].copy() if 'comment' in df1.columns else df1[['posts']].copy()
df1_text.rename(columns={df1_text.columns[0]: 'text'}, inplace=True)
df1_text['source'] = 'instagram_comments'
df1_text['label'] = 'netral'  # diasumsikan tanpa label

# TikTok 2025-05-18
df2_text = df2[['text']].copy()
df2_text['source'] = 'tiktok_2025_05_18'
df2_text['label'] = 'netral'

# TikTok 2025-05-15
df3_text = df3[['text']].copy()
df3_text['source'] = 'tiktok_2025_05_15'
df3_text['label'] = 'netral'

# Twitter: kolom gabungan 'sentimen\tTweet' → pecah jadi dua kolom
df4[['label', 'text']] = df4['sentimen\tTweet'].str.split('\t', expand=True)
df4['source'] = 'twitter_sentiment'
df4_text = df4[['text', 'label', 'source']].copy()

# Review Jabar: ubah skor ke label
df5_text = df5[['content', 'score']].copy()
df5_text.rename(columns={'content': 'text'}, inplace=True)
df5_text['source'] = 'review_layanan_jabar'

def score_to_label(score):
    if score <= 2:
        return 'negatif'
    elif score == 3:
        return 'netral'
    else:
        return 'positif'

df5_text['label'] = df5_text['score'].apply(score_to_label)
df5_text.drop(columns=['score'], inplace=True)
```

```python
df_all = pd.concat([df1_text, df2_text, df3_text, df4_text, df5_text], ignore_index=True)
```

```python
# Cek hasil
print(df_all.head())
print(df_all['label'].value_counts())
print(df_all['source'].value_counts())
```

```python
df_all.to_csv("combined_dataset.csv", index=False)
```
