```python
def get_rekomendasi(jenis, karakteristik, sumber):
    jenis = jenis.lower()
    karakteristik = karakteristik.lower()
    sumber = sumber.lower()

    # Telusuri kategori sampah
    for kategori_item in struktur_narasi_sampah:
        if kategori_item["kategori"].lower() == jenis:
            for karakteristik_item in kategori_item["karakteristik_list"]:
                if karakteristik_item["karakteristik"].lower() == karakteristik:
                    for sumber_item in karakteristik_item["sumber_dan_dampak"]:
                        if sumber in sumber_item["sumber"].lower():
                            # Rekomendasi berbasis sumber dan karakteristik
                            if jenis == "organik":
                                return (
                                    f"Sampah dari {sumber_item['sumber']} yang memiliki karakteristik {karakteristik} "
                                    "sebaiknya diolah menjadi kompos yang dapat meningkatkan kesuburan tanah. "
                                    "Lakukan pemilahan sejak awal dan simpan dalam wadah tertutup agar proses pengomposan berjalan optimal dan mengurangi bau tidak sedap."
                                )
                            elif jenis == "anorganik":
                                if "daur ulang" in karakteristik:
                                    return (
                                        f"Pisahkan sampah dari {sumber_item['sumber']} dan pastikan dalam kondisi kering serta bersih. "
                                        "Serahkan sampah tersebut ke bank sampah atau pusat daur ulang terdekat untuk mengurangi volume sampah di TPA dan mendukung ekonomi sirkular."
                                    )
                                elif "mikroplastik" in karakteristik:
                                    return (
                                        f"Kurangi penggunaan plastik dari {sumber_item['sumber']} dan kumpulkan dalam satu tempat khusus. "
                                        "Bawa sampah plastik ini ke fasilitas pengolahan sampah plastik yang mampu mencegah pecah menjadi mikroplastik yang berbahaya bagi lingkungan dan kesehatan."
                                    )
                                else:
                                    return (
                                        f"Kumpulkan sampah dari {sumber_item['sumber']} secara terpisah dan hindari pembakaran terbuka. "
                                        "Pisahkan sesuai jenisnya, seperti plastik, logam, dan kaca, untuk memudahkan proses daur ulang dan mengurangi polusi udara."
                                    )
                            elif jenis == "b3":
                                if "beracun" in karakteristik or "berbahaya" in karakteristik:
                                    return (
                                        f"Limbah dari {sumber_item['sumber']} termasuk kategori B3 yang beracun dan berbahaya. "
                                        "Harus dipisahkan dari sampah biasa, disimpan dalam wadah berlabel khusus, dan ditangani oleh dinas lingkungan atau fasilitas pengelola limbah resmi agar tidak mencemari lingkungan dan membahayakan kesehatan."
                                    )
                                elif "bisa bernilai" in karakteristik:
                                    return (
                                        f"Sampah dari {sumber_item['sumber']} memiliki nilai ekonomi tinggi. "
                                        "Pisahkan dan serahkan ke pengelola limbah B3 berizin yang dapat mendaur ulang logam atau plastik elektronik, sehingga limbah dapat diolah kembali dan mengurangi eksploitasi sumber daya alam."
                                    )
                                else:
                                    return (
                                        f"Limbah dari {sumber_item['sumber']} memerlukan penanganan khusus sesuai karakteristik B3. "
                                        "Jangan dicampur dengan limbah lain dan hindari kontak langsung agar risiko pencemaran dan bahaya kesehatan dapat diminimalisir."
                                    )
    # Fallback jika tidak ditemukan
    return "Rekomendasi tidak ditemukan. Periksa kembali jenis, karakteristik, dan sumber yang dimasukkan."
```

```python
# Konversi struktur_narasi_sampah ke struktur_sampah yang mudah diakses
struktur_sampah = {}
for kategori_item in struktur_narasi_sampah:
    kategori = kategori_item["kategori"].lower()
    entri_list = []
    for karakteristik_item in kategori_item["karakteristik_list"]:
        karakteristik = karakteristik_item["karakteristik"].lower()
        for sumber_item in karakteristik_item["sumber_dan_dampak"]:
            entri = {
                "sumber": sumber_item["sumber"],
                "karakteristik": [karakteristik],
                "dampak": sumber_item.get("dampak", [])
            }
            entri_list.append(entri)
    struktur_sampah[kategori] = entri_list

# Daftar kategori sampah
kategori_sampah = list(struktur_sampah.keys())
```

```python
import time
import random

# Generate dataset narasi kebijakan sampah seimbang
output_path = '/home/jupyter/data/raw/llm_texts_raw/dataset_narasi_kebijakan_sampah_fix.txt'
total_data = 500
data_per_kategori = total_data // len(kategori_sampah)

start_time = time.time()

with open(output_path, 'w', encoding='utf-8') as f:
    for jenis in kategori_sampah:
        entri_list = struktur_sampah[jenis]

        for _ in range(data_per_kategori):
            entri = random.choice(entri_list)

            sumber = entri["sumber"]
            karakteristik = entri["karakteristik"][0]  # karena list 1 elemen
            dampak = entri["dampak"]

            dipilih_kar = karakteristik  # ambil karakteristik utama
            dampak_terpilih = random.sample(dampak, k=min(2, len(dampak)))
            narasi_rekomendasi = get_rekomendasi(jenis, karakteristik, sumber)

            narasi_kesimpulan = (
                f"Jenis sampah yang teridentifikasi adalah kategori {jenis}. "
                f"Karakteristik utamanya adalah {dipilih_kar}. Umumnya berasal dari {sumber}. "
                f"Dampak negatif yang mungkin timbul antara lain: {', '.join(dampak_terpilih)}."
            )
            final_narasi = (
                f"Kesimpulan: {narasi_kesimpulan}\n"
                f"Rekomendasi Penanganan: {narasi_rekomendasi}\n\n"
            )

            f.write(final_narasi)

print("Dataset narasi kebijakan sampah berhasil digenerate dan tersimpan.")
print(f"Waktu eksekusi: {time.time() - start_time:.2f} detik")
```
